```python
#!pip install pytube
from pytube import YouTube
#!pip install pytesseract
import pytesseract
import cv2 as cv
#from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import re
```

```python
#Change to any URL
url = "https://youtu.be/zk5ophL7wBo"
video = YouTube(url)
filename = re.sub('[^a-zA-Z0-9]', '_', video.title)
#Run this to download the youtube video
#stream = video.streams.filter(res="360p").first().download(filename = filename)
```

```python
details = video.streams.filter(res="360p").first()
print(details)
```

```python
cap = cv.VideoCapture(filename+'.mp4');
cap.set(cv.CAP_PROP_FPS, details.fps)

#testing
if not cap.isOpened():
  print('Video is not set')
```

```python
exp  = re.compile('\S+')
prev = ''

while cap.isOpened():

    ret, frame = cap.read()
    #frame = cv.cvtColor()

    cv.imshow('Video', frame)

    text = pytesseract.image_to_string(frame, lang = 'eng')
    
    if len(text) and exp.search(text) is not None and text != prev:
        print(text)
        
    prev = text

    if cv.waitKey(1) & 0xFF == ord('q'):  #Enter q to quit
        break

cv.destroyAllWindows()
```
